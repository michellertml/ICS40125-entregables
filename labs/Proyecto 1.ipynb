Problema 1

```python
def calcular_pi(n):
    suma_serie = 0.0

    # El ciclo empieza en 1 y termina en n (por eso usamos n + 1)
    # Esto replica exactamente el 'k=1' hasta 'n' de tu sumatoria
    for k in range(1, n + 1):
        numerador = (-1) ** (k + 1)
        denominador = 2 * k - 1
        suma_serie += numerador / denominador

    return 4 * suma_serie

# --- Zona de pruebas para Google Colab ---
print(f"Resultado con 3 términos: {calcular_pi(3)}")
print(f"Resultado con 1000 términos: {calcular_pi(1000)}")
```

Problema 2


```python
# --- Parte A: Función para calcular el factorial ---
def factorial(n):
    # El factorial de 0 y 1 es siempre 1
    if n == 0 or n == 1:
        return 1

    resultado = 1
    # Multiplicamos desde 2 hasta n
    for i in range(2, n + 1):
        resultado *= i
    return resultado


# --- Parte B: Función para aproximar el número e ---
def calcular_e(n):
    suma_serie = 0.0

    # El ciclo genera los primeros 'n' términos de la sumatoria
    # Como k empieza en 0, range(n) va exactamente desde 0 hasta n-1
    for k in range(n):
        # Usamos nuestra función factorial(k) creada en la Parte A
        termino = 1 / factorial(k)
        suma_serie += termino

    return suma_serie


# --- Zona de pruebas para verificar en Google Colab ---
print(f"Aproximación de e con 1 término:  {calcular_e(1)}")   # Debe dar 1.0 (ya que 1 / 0! = 1/1 = 1)
print(f"Aproximación de e con 3 términos: {calcular_e(3)}")   # Debe dar 2.5 (1 + 1/1! + 1/2!)
print(f"Aproximación de e con 10 términos:{calcular_e(10)}")  # Se acerca mucho al valor real (~2.71828)
```

Problema 3

```python
# --- Parte A: Función para obtener la lista de divisores propios ---
def divisores_propios(n):
    lista_divisores = []

    # Recorremos desde 1 hasta n-1 (porque el número n no se incluye)
    for i in range(1, n):
        # Si el residuo de la división es 0, es un divisor propio
        if n % i == 0:
            lista_divisores.append(i)

    return lista_divisores


# --- Parte B: Función para verificar si dos números son amigos ---
def amigos(n1, n2):
    # Calculamos la suma de los divisores propios de n1 usando sum()
    suma_n1 = sum(divisores_propios(n1))

    # Calculamos la suma de los divisores propios de n2
    suma_n2 = sum(divisores_propios(n2))

    # Retornamos True si se cumple la condición cruzada, de lo contrario False
    # Condición: sigma(n1) == n2 Y sigma(n2) == n1
    if suma_n1 == n2 and suma_n2 == n1:
        return True
    else:
        return False


# --- Zona de pruebas para verificar en Google Colab ---
# Caso clásico del enunciado: 220 y 284
print(f"¿220 y 284 son amigos?: {amigos(220, 284)}")  # Debe dar True

# Prueba con otros números que NO son amigos
print(f"¿100 y 150 son amigos?: {amigos(100, 150)}")  # Debe dar False

# Otro par famoso de números amigos para comprobar: 1184 y 1210
print(f"¿1184 y 1210 son amigos?: {amigos(1184, 1210)}")  # Debe dar True
```

Problema 4

```python
def collatz(n):
    # Inicializamos la lista de la secuencia incluyendo el número inicial
    secuencia = [n]

    # El ciclo continúa hasta que n se convierta en 1
    while n > 1:
        # Si el número es par
        if n % 2 == 0:
            n = n // 2  # Usamos división entera (//) para mantener números enteros
        # Si el número es impar
        else:
            n = 3 * n + 1

        # Agregamos el nuevo valor generado a nuestra lista
        secuencia.append(n)

    return secuencia


# --- Zona de pruebas para verificar en Google Colab ---
# Prueba con n = 6
print(f"Secuencia para 6: {collatz(6)}")
# Resultado esperado: [6, 3, 10, 5, 16, 8, 4, 2, 1]

# Prueba con n = 11
print(f"Secuencia para 11: {collatz(11)}")
# Resultado esperado: [11, 34, 17, 52, 26, 13, 40, 20, 10, 5, 16, 8, 4, 2, 1]

# Prueba con n = 1 (caso base, debería retornar solo [1])
print(f"Secuencia para 1: {collatz(1)}")
```

Problema 5

```python
# --- Parte A: Función para verificar si un número es primo ---
def es_primo(num):
    # Los números menores o iguales a 1 no son primos
    if num <= 1:
        return False

    # Verificamos si tiene algún divisor desde 2 hasta la raíz cuadrada de num (o num-1)
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False  # Si encuentra un divisor, no es primo

    return True  # Si no encontró ningún divisor, es primo


# --- Parte B: Función para generar la lista de primos hasta n ---
def lista_de_primos(n):
    primos = []
    # Recorremos todos los números desde 2 hasta n (inclusive)
    for i in range(2, n + 1):
        # Si el número actual es primo usando la función de la Parte A, lo guardamos
        if es_primo(i):
            primos.append(i)
    return primos


# --- Parte C: Función para encontrar los dos primos que suman n ---
def goldbach(n):
    # 1. Obtenemos todos los primos candidatos hasta n
    primos_disponibles = lista_de_primos(n)

    # 2. Buscamos la combinación de dos primos que sumen n
    for n1 in primos_disponibles:
        for n2 in primos_disponibles:
            if n1 + n2 == n:
                # Retornamos el primer par encontrado como una tupla (n1, n2)
                return n1, n2


# --- Zona de pruebas para verificar en Google Colab ---
print(f"Goldbach para 4:   {goldbach(4)}")    # Esperado: (2, 2) -> 2 + 2 = 4
print(f"Goldbach para 10:  {goldbach(10)}")   # Esperado: (3, 7) o (5, 5) -> 3 + 7 = 10
print(f"Goldbach para 28:  {goldbach(28)}")   # Esperado: (5, 23) o (11, 17) -> 5 + 23 = 28
print(f"Goldbach para 100: {goldbach(100)}")  # Esperado: (3, 97) o similar
```
