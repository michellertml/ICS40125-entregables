```python
import pandas as pd

# Cargar datos
df = pd.read_csv('https://raw.githubusercontent.com/fralfaro/ICS40125/main/docs/labs/data/netflix_titles.csv')
df.head()
```

```python
# ¿Cuántas filas y columnas tiene?
print(f"Dimensiones del dataset: {df.shape}")

# ¿Qué tipos de datos hay y cuántos nulos por columna?
print("\nInformación del DataFrame:")
print(df.info())

# Conteo directo de nulos por columna
print("\nValores nulos por columna:")
print(df.isnull().sum())
```

```python
# 1. Convertimos y asignamos en silencio
df['date_added'] = pd.to_datetime(df['date_added'].astype(str).str.strip(), errors='coerce')

df = df.assign(
    year_added=df['date_added'].dt.year,
    month_added=df['date_added'].dt.strftime('%B')
)

# Despliega solo las columnas de fecha para verificar el éxito
df[['title', 'date_added', 'year_added', 'month_added']].head()
```

parte 2


```python
peliculas_post_2018 = df.loc[(df['type'] == 'Movie') & (df['year_added'] > 2018)]

# Mostramos cuántas encontramos y las primeras filas
print(f"Películas encontradas: {len(peliculas_post_2018)}")
peliculas_post_2018[['title', 'type', 'year_added']].head()
```

```python
# 1. Filtrar títulos que contienen 'love'
titulos_love = df.loc[df['title'].str.contains('love', case=False, na=False)]
print(f"Títulos que contienen 'love': {len(titulos_love)}")

# 2. Extraer la duración en minutos para las películas
df_movies = df.loc[df['type'] == 'Movie'].copy()
df_movies['duration_num'] = df_movies['duration'].str.extract(r'(\d+)').astype(float)

# Verificamos que se extrajo correctamente el número
df_movies[['title', 'duration', 'duration_num']].head()
```

```python
# Convertimos el texto separado por comas en una lista real
df_exploded = df.copy()
df_exploded['listed_in'] = df_exploded['listed_in'].str.split(', ')

# Explotamos la columna
df_exploded = df_exploded.explode('listed_in')

# Obtenemos el Top 10 de géneros más frecuentes
top_10_generos = df_exploded['listed_in'].value_counts().head(10)

print("--- TOP 10 GÉNEROS MÁS FRECUENTES ---")
print(top_10_generos)
```

```python
# Convertimos el texto separado por comas en una lista real
df_exploded = df.copy()
df_exploded['listed_in'] = df_exploded['listed_in'].str.split(', ')

# Explotamos la columna
df_exploded = df_exploded.explode('listed_in')

# Obtenemos el Top 10 de géneros más frecuentes
top_10_generos = df_exploded['listed_in'].value_counts().head(10)

print("--- TOP 10 GÉNEROS MÁS FRECUENTES ---")
print(top_10_generos)
```

```python
peliculas_especificas = df_movies.loc[
    (df_movies['duration_num'] > 100) &
    (df_movies['rating'] == 'R') &
    (df_movies['country'] == 'United States')
]

print(f"Se encontraron {len(peliculas_especificas)} películas con estos criterios.")
peliculas_especificas[['title', 'duration_num', 'rating', 'country']].head()
```

```python
top_10_largas = df_movies.nlargest(10, 'duration_num')[['title', 'duration_num']]

# Esto generará una tabla estilizada con barras de color en tu notebook
top_10_largas.style.background_gradient(cmap='Reds', subset=['duration_num'])
```

```python
print("=== RESPUESTA DESAFÍO: Combinaciones más frecuentes ===")
# Usamos el dataframe explotado que creamos en el paso de géneros
combinaciones = df_exploded.value_counts(subset=["listed_in", "rating"]).head(10)
print(combinaciones)

print("\n=== RESPUESTA BONUS: Duplicados ===")
# 1. Títulos repetidos pero con distinto año de lanzamiento (ordenados para comparar)
duplicados_titulo = df[df.duplicated(subset=['title'], keep=False)].sort_values(by='title')
print(f"Filas con títulos repetidos detectadas: {len(duplicados_titulo)}")

# 2. Total de títulos únicos
titulos_unicos = df['title'].nunique()
print(f"Cantidad total de títulos únicos en el catálogo: {titulos_unicos}")
```
